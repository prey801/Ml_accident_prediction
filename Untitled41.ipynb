```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import chardet

def detect_file_encoding(file_path):
    """
    Detect the encoding of a file using chardet
    """
    with open(file_path, 'rb') as file:
        raw_data = file.read()
    result = chardet.detect(raw_data)
    return result['encoding']

def load_data(file_path):
    """
    Load the CSV file with appropriate encoding
    """
    try:
        # First try UTF-8
        print("Attempting to read with UTF-8 encoding...")
        return pd.read_csv(file_path)
    except UnicodeDecodeError:
        print("UTF-8 encoding failed. Detecting file encoding...")
        # Detect the file encoding
        encoding = detect_file_encoding(file_path)
        print(f"Detected encoding: {encoding}")

        # Try reading with detected encoding
        try:
            return pd.read_csv(file_path, encoding=encoding)
        except:
            print("Detected encoding failed. Trying common encodings...")
            # Try common encodings
            encodings = ['latin1', 'iso-8859-1', 'cp1252']
            for enc in encodings:
                try:
                    print(f"Attempting to read with {enc} encoding...")
                    df = pd.read_csv(file_path, encoding=enc)
                    print(f"Successfully read file with {enc} encoding")
                    return df
                except:
                    continue
            raise ValueError("Unable to read file with any common encoding")

def prepare_data(data):
    """
    Prepare and clean the data for modeling.
    """
    try:
        # Create a copy to avoid modifying original data
        df = data.copy()

        print("\nInitial data shape:", df.shape)
        print("\nColumns in dataset:")
        for col in df.columns:
            print(f"- {col}: {df[col].nunique()} unique values")
            print(f"  Sample values: {df[col].unique()[:5]}")

        # Create severity score mapping
        severity_mapping = {
            'No injury/unknown': 0,
            'Non-incapacitating': 1,
            'Incapacitating': 2
        }

        # Validate that required columns exist
        required_columns = ['Weekend?', 'Hour', 'Collision Type', 'Injury Type', 'Primary Factor']
        missing_columns = [col for col in required_columns if col not in df.columns]
        if missing_columns:
            raise ValueError(f"Missing required columns: {missing_columns}")

        # Convert Hour to numeric format
        if df['Hour'].dtype == object:  # If hour is string
            # Remove any non-numeric characters and convert to numeric
            df['Hour'] = df['Hour'].astype(str).str.extract('(\d+)').astype(float)

        # Handle missing values
        df['Hour'] = df['Hour'].fillna(df['Hour'].median())
        df['Primary Factor'] = df['Primary Factor'].fillna('Unknown')
        df['Collision Type'] = df['Collision Type'].fillna('Unknown')

        # Create binary features
        df['is_weekend'] = (df['Weekend?'] == 'Weekend').astype(int)
        df['is_multiple_vehicles'] = (df['Collision Type'].isin(['2-Car', '3+ Cars'])).astype(int)

        # Create severity score as dependent variable
        df['severity_score'] = df['Injury Type'].map(severity_mapping)

        # Handle any unmapped severity scores
        if df['severity_score'].isnull().any():
            print(f"Warning: Found {df['severity_score'].isnull().sum()} unknown injury types")
            print("Unique injury types:", df['Injury Type'].unique())
            df['severity_score'] = df['severity_score'].fillna(0)  # Default to no injury

        # Create dummy variables for categorical columns
        primary_factor_dummies = pd.get_dummies(df['Primary Factor'], prefix='factor')
        collision_type_dummies = pd.get_dummies(df['Collision Type'], prefix='collision')

        # Combine features
        features = pd.concat([
            df[['Hour', 'is_weekend', 'is_multiple_vehicles']],
            primary_factor_dummies,
            collision_type_dummies
        ], axis=1)

        print("\nFinal feature matrix shape:", features.shape)
        print("\nFeature names:")
        for i, col in enumerate(features.columns, 1):
            print(f"{i}. {col}")

        return features, df['severity_score']

    except Exception as e:
        print(f"Error in data preparation: {str(e)}")
        print("DataFrame info:")
        print(df.info())
        raise

# Main execution
if __name__ == "__main__":
    try:
        file_path = '/content/monroe county car crach 2003-2015.csv'
        print("Loading data...")

        # Load data with appropriate encoding
        df = load_data(file_path)

        # Continue with the rest of the analysis...
        print("\nPreparing data...")
        X, y = prepare_data(df)

        # Train the model
        print("\nTraining model...")
        model, X_train, X_test, y_test, y_pred, mse, r2 = train_model(X, y)

        # Save the model
        print("\nSaving model...")
        with open('accident_severity_model.pkl', 'wb') as file:
            pickle.dump(model, file)

        # Print model performance
        print("\nModel Performance:")
        print(f"Mean Squared Error: {mse:.4f}")
        print(f"RÂ² Score: {r2:.4f}")

        # Plot results
        plt.figure(figsize=(10, 6))
        plt.scatter(y_test, y_pred, alpha=0.5)
        plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
        plt.xlabel('Actual Severity')
        plt.ylabel('Predicted Severity')
        plt.title('Actual vs Predicted Accident Severity')
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"Error in main execution: {str(e)}")
```

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap
import seaborn as sns
import matplotlib.pyplot as plt

def create_accident_heatmap(file_path):
    """
    Create both an interactive Folium heatmap and a static seaborn heatmap
    of accident locations
    """
    try:
        # Read the data
        df = pd.read_csv(file_path, encoding='latin1')

        # Remove rows where latitude or longitude is 0 or null
        df = df[df['Latitude'] != 0]
        df = df[df['Longitude'] != 0]
        df = df.dropna(subset=['Latitude', 'Longitude'])

        print(f"Total accidents with valid coordinates: {len(df)}")

        # Create interactive Folium heatmap
        print("Creating interactive heatmap...")

        # Calculate the center point for the map
        center_lat = df['Latitude'].mean()
        center_lon = df['Longitude'].mean()

        # Create base map
        m = folium.Map(location=[center_lat, center_lon],
                      zoom_start=12,
                      tiles='CartoDB positron')

        # Prepare data for heatmap
        locations = df[['Latitude', 'Longitude']].values.tolist()

        # Add heatmap layer
        HeatMap(locations,
                radius=15,
                blur=10,
                max_zoom=13).add_to(m)

        # Save interactive map
        m.save('accident_heatmap_interactive.html')
        print("Interactive heatmap saved as 'accident_heatmap_interactive.html'")

        # Create static heatmap using seaborn
        print("Creating static heatmap...")

        plt.figure(figsize=(15, 10))

        # Create 2D histogram
        heatmap_data, xedges, yedges = np.histogram2d(
            df['Longitude'],
            df['Latitude'],
            bins=50
        )

        # Create heatmap using seaborn
        sns.heatmap(heatmap_data.T,
                   cmap='YlOrRd',
                   cbar_kws={'label': 'Number of Accidents'})

        plt.title('Accident Density Heatmap')
        plt.xlabel('Longitude')
        plt.ylabel('Latitude')

        # Add detailed statistics
        print("\nLocation Statistics:")
        print(f"Latitude range: {df['Latitude'].min():.4f} to {df['Latitude'].max():.4f}")
        print(f"Longitude range: {df['Longitude'].min():.4f} to {df['Longitude'].max():.4f}")

        # Create time-based heatmaps
        print("\nCreating time-based analysis...")

        # Hour-based heatmap
        plt.figure(figsize=(15, 5))
        hour_accidents = df['Hour'].value_counts().sort_index()
        sns.heatmap([hour_accidents.values],
                   cmap='YlOrRd',
                   cbar_kws={'label': 'Number of Accidents'})
        plt.title('Accidents by Hour of Day')
        plt.xlabel('Hour')
        plt.yticks([])

        # Weekend vs Hour heatmap
        plt.figure(figsize=(15, 5))
        weekend_hour_data = pd.crosstab(df['Weekend?'], df['Hour'])
        sns.heatmap(weekend_hour_data,
                   cmap='YlOrRd',
                   cbar_kws={'label': 'Number of Accidents'})
        plt.title('Accidents by Hour and Day Type')
        plt.xlabel('Hour')
        plt.ylabel('Day Type')

        # Additional analysis: Accident types by location
        print("\nAnalyzing accident patterns...")


        # Save the figures
        plt.savefig('accident_heatmaps.png')
        print("Static heatmaps saved as 'accident_heatmaps.png'")

        return {
            'total_accidents': len(df),
            'accident_types': df['Collision Type'].value_counts().to_dict(),
            'peak_hour': hour_accidents.idxmax(),
            'peak_hour_count': hour_accidents.max(),
            'coord_ranges': {
                'lat_range': (df['Latitude'].min(), df['Latitude'].max()),
                'lon_range': (df['Longitude'].min(), df['Longitude'].max())
            }
        }

    except Exception as e:
        print(f"Error creating heatmap: {str(e)}")
        raise

if __name__ == "__main__":
    try:
        file_path = '/content/monroe county car crach 2003-2015.csv'
        stats = create_accident_heatmap(file_path)

        print("\nSummary Statistics:")
        print(f"Total accidents analyzed: {stats['total_accidents']}")
        print(f"\nAccident types distribution:")
        for type_, count in stats['accident_types'].items():
            print(f"- {type_}: {count}")
        print(f"\nPeak accident hour: {stats['peak_hour']} (with {stats['peak_hour_count']} accidents)")

    except Exception as e:
        print(f"Error in main execution: {str(e)}")
```
